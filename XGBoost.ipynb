**Cross validation strategy selected:**
Due to the fact that we have imbalance classes, the cross validation strategy which was selected for the model evaluation is StratifiedShuffleSplit. Stratified returns stratified randomized folds. In stratified k-fold cross-validation, the folds are selected so that the mean response value is approximately equal in all the folds. In the case of a dichotomous classification, this means that each fold contains roughly the same proportions of the two types of class labels.

### 2.4) Various models and encoding technique analysis

In order to compare the results of each aforementioned categorical features techniques (See chapter 2.2 above - "Dealing with categorical features") it was decided to define model without any categorical features as a baseline result

```python
def compare_encoding_techniques(data, YLFP, test_size, seed):
    "Function calculates the accuracy on the cross validation and test data and prints the results"
    
    "Data without categorical features"
    X1=data.loc[:, data.dtypes != 'object']
    
    # Spliting data to train and test data:
    X_train1, X_test1, y_train1, y_test1 = model_selection.train_test_split(X1, YLFP, test_size=test_size,random_state=seed)
    
    # Accuracy on cross validation:
    print("Data with no categorical features - Accuracy on the cross validation is:", 
    round(np.mean(model_selection.cross_val_score(xgboost, X_train1, y_train1, cv=skf, scoring='accuracy')), 4))
    
    # Fit the model on the train data:
    model1=xgboost.fit(X_train1, y_train1)
    
    # Accuracy on test data:
    print("Data with no categorical features - Accuracy on test data is:", 
    round (metrics.accuracy_score(y_test1, model1.predict(X_test1)),4))
    print("-------------------------------------------------------------------------------")
    
    
    "One Hot encoding"
    cols = list(data.select_dtypes(include=['object']).columns)
    X2 = pd.get_dummies(data=data, columns=cols,sparse=True)
    
    # Spliting data to train and test for the model:
    X_train, X_test, y_train, y_test = model_selection.train_test_split(np.array(X2), YLFP, test_size=test_size,
                                                                        random_state=seed)
    # Accuracy on cross validation:
    print("One Hot Encoding - Accuracy on the cross validation is:", 
    round(np.mean(model_selection.cross_val_score(xgboost, X_train, y_train, cv=skf, scoring='accuracy')),4))
    
    # Fit the model on the train data:
    model2=xgboost.fit(X_train, y_train)
    
    # Accuracy on test data:
    print("One Hot Encoding - Accuracy on test data is:", 
    round (metrics.accuracy_score(y_test, model2.predict(X_test)),4))
    print("-------------------------------------------------------------------------------")
    

    "Mean target technique"
    X=data
    train_df_part, valid_df, y_train_part, y_valid = model_selection.train_test_split(X, YLFP, test_size=test_size,
                                                                                  random_state=seed)
    
    "Function calculates a mean for every category of categorical feature using a cross validation"
    def mean_target_enc(train_df, y_train, valid_df):   
        glob_mean = y_train.mean()
        train_df = pd.concat([train_df, pd.Series(y_train, name='y')], axis=1)
        new_train_df = train_df.copy()
    
        cat_features = train_df.columns[train_df.dtypes == 'object'].tolist()    

        for col in cat_features:
            new_train_df[col + '_mean_target'] = [glob_mean for _ in range(new_train_df.shape[0])]

        for train_idx, valid_idx in skf.split(train_df, y_train):
            train_df_cv, valid_df_cv = train_df.iloc[train_idx, :], train_df.iloc[valid_idx, :]

            for col in cat_features:
                means = valid_df_cv[col].map(train_df_cv.groupby(col)['y'].mean())
                valid_df_cv[col + '_mean_target'] = means.fillna(glob_mean)
            
            new_train_df.iloc[valid_idx] = valid_df_cv
    
        new_train_df.drop(cat_features + ['y'], axis=1, inplace=True)
    
        for col in cat_features:
            means = valid_df[col].map(train_df.groupby(col)['y'].mean())
            valid_df[col + '_mean_target'] = means.fillna(glob_mean)
        
        valid_df.drop(train_df.columns[train_df.dtypes == 'object'], axis=index_column, inplace=True)
    
        return new_train_df, valid_df

    train_mean_target_part, valid_mean_target = mean_target_enc(train_df_part, y_train_part, valid_df)
    
    # Accuracy on cross validation:
    print("Mean target Encoding - Accuracy on the cross validation is:", 
    round(np.mean(model_selection.cross_val_score(xgboost, train_mean_target_part, y_train_part, cv=skf, 
                                                  scoring='accuracy')),4))
    
    # Fit the model on the train data:
    model3=xgboost.fit(train_mean_target_part, y_train_part)
    
    # Accuracy on test data:
    print("Mean target Encoding - Accuracy on test data is:", 
    round(metrics.accuracy_score(y_valid, model3.predict(valid_mean_target)),4))
    print("-------------------------------------------------------------------------------")
        
    "Catboost implementation"    
    ctb = CatBoostClassifier(random_seed=seed)
    X=data
    YLFP=YLFP.astype(int)
    X=X.fillna(999999)
    train_df_part, valid_df, y_train_part, y_valid = model_selection.train_test_split(X, YLFP, test_size=test_size,
                                                                                      random_state=seed)
    # Defining all categorical features:
    cat_features_idx = np.where(train_df_part.dtypes == 'object')[0].tolist()
    
    # Accuracy on cross validation:
    cv_scores = []
    for train_idx, test_idx in skf.split(train_df_part, y_train_part):
        cv_train_df, cv_valid_df = train_df_part.iloc[train_idx, :], train_df_part.iloc[test_idx, :]
        y_cv_train, y_cv_valid = y_train_part.iloc[train_idx], y_train_part.iloc[test_idx]
    
        ctb.fit(cv_train_df, y_cv_train,
            cat_features=cat_features_idx, silent=True);
        cv_scores.append(metrics.accuracy_score(y_cv_valid, ctb.predict(cv_valid_df)))
        
    # Accuracy on cross validation:
    print("Catboost - Accuracy on the cross validation is:", round(np.mean(cv_scores),4))
    
    # Fit the model on the train data:
    model4=ctb.fit(train_df_part, y_train_part, cat_features=cat_features_idx, silent=True)
    
    # Accuracy on test data:
    print("Catboost - Accuracy on test data is:", round(metrics.accuracy_score(y_valid, model4.predict(valid_df)),4))
```

### 3.1) Models evaluation prior implementing a cross validation

In order to verify and check the effectiveness of XGBoost model, it was decided to compared it with the other models (Random forest, Decision trees and Logistic regression). All algorithms parameters were set to the default settings. The idea was to check which algorithm works better with default parameters "out of the box".

```python
def train_and_predict_models(X_train, y_train, X_test, seed):
    X_train=X_train.fillna(999999) # DecisionTree and Random forest can't obtain NaN values
    y_train=y_train.fillna(999999) # DecisionTree and Random forest can't obtain NaN values
    X_test=X_test.fillna(999999) # DecisionTree and Random forest can't obtain NaN values
    
    # Train the models:
    #1. Decision Trees 
    decision_tree = tree.DecisionTreeClassifier(random_state=seed)
    decision_tree.fit(X_train, y_train)

    #2. Random Forest
    random_forest = ensemble.RandomForestClassifier(random_state=seed)
    random_forest.fit(X_train, y_train)  

    #3. Logistic Regression
    logistic_reg = LogisticRegression(random_state=seed)
    logistic_reg.fit(X_train, y_train)

    #4 XGboost
    gbc = ensemble.GradientBoostingClassifier(random_state=seed)
    gbc.fit(X_train, y_train)
    
    # Predict the labels output and probabilities for each predicted class using the selected models:
    #1 Decision Trees
    decision_tree_predict = decision_tree.predict(X_test)
    decision_tree_predict_prob = decision_tree.predict_proba(X_test)

    #2 Random Forrest
    random_forest_predict = random_forest.predict(X_test)
    random_forest_predict_prob = random_forest.predict_proba(X_test)

    #4 Logistic Regression
    logistic_reg_predict = logistic_reg.predict(X_test)
    logistic_reg_predict_prob = logistic_reg.predict(X_test)

    #3 XGboost
    gbc_y_predict = gbc.predict(X_test)
    gbc_y_pred_prob = gbc.predict_proba(X_test) 
    
    return decision_tree_predict, decision_tree_probability, random_forest_predict, random_forest_predict_probility, 
           logistic_reg_predict, logistic_reg_predict_prob, gbc_y_predict, gbc_y_pred_prob
```

### 3.2) Models evaluation using a cross validation

Taking in the consideration the fact that we are dealing with the classification problem (LFP prediction), the following metrics were selected in order to compare the models side by side using a cross validation:
- Accuracy - Measures the classification model accuracy as the proportion of true results to total cases.
- Precision - The proportion of true results over all positive results.
- Recall - The fraction of all correct results returned by the model.
- F-score - Computed as the weighted average of precision and recall between 0 and 1, where the ideal F-score value is 1.
- AUC - Measures the area under the curve plotted with true positives on the y axis and false positives on the x axis.
- Log loss - A single score used to express the penalty for wrong results. It is calculated as the difference between two probability distributions – the true one, and the one in the model.

```python
# Define a list of models and metrics to evaluate:
methods=['accuracy','recall','precision','f1','log_loss','roc_auc']
models=['decision_tree','random_forest','logistic_regression','gbc']

# Cross validation strategy defenition:
cv_strategy = model_selection.StratifiedShuffleSplit(n_splits = folds ,test_size = test_size, random_state = seed)
```

```python
def cross_val_score (X, y, seed, methods, models, cv_strategy):
    "Function evaluates model performance using a Stratified Cross Validation for the selected metrics:"
    
    decision_trees=[]
    scores=[]
    random_forests=[]
    logistic_regression=[]
    gxboost=[]
    for model in models:
        if model=='decision_tree':
            for method in methods:     
                scoring = cross_validation.cross_val_score(decision_tree, X, y, scoring = method, cv = cv_strategy)
                k=np.mean(scoring)
                decision_trees.append(k)
        elif model=='random_forest':
            for method in methods:   
                scoring = cross_validation.cross_val_score(random_forest, X, y, scoring = method, cv = cv_strategy)
                k=np.mean(scoring)
                random_forests.append(k)
        elif model=='logistic_regression':
            for method in methods:   
                scoring = cross_validation.cross_val_score(logistic_reg, X, y, scoring = method, cv = cv_strategy)
                k=np.mean(scoring)
                logistic_regression.append(k)
        else:
            for method in methods:   
                scoring = cross_validation.cross_val_score(gbc, X, y, scoring = method, cv = cv_strategy)
                k=np.mean(scoring)
                gxboost.append(k)            
    return decision_trees, random_forests, logistic_regression, gxboost
```

### 4.1) Train data errors vs Cross validation errors - Bias vs Variance trade off

**Bias** - Bias is the difference between our model prediction and the correct value which we are trying to predict. Model with high bias pays very little attention to the training data and oversimplifies the model. It always leads to high error on training and test data.

**Variance** - Variance is the variability of model prediction for a given data point or a value which tells us spread of our data. Model with high variance pays a lot of attention to training data and sometimes does not generalize on the data which it hasn’t seen before. As a result, such models perform very well on training data but has a potential of higher error rates on test data.

In order to identify if we are dealing with high variance or high bias problem we built a graph of model error as a function of training set size for train data and test data (CV).

```python
def Errors_vs_data_sizes (trees, score_method, seed, train_sizes, X=data, y=YLFP):
    "Function plots log loss error rates on the cross validation and the train data:"
    
    estimator = XGBClassifier(max_depth=trees_depth,n_estimators=trees,warm_start=True,random_state=seed)
    train_sizes, train_scores, test_scores = model_selection.learning_curve(estimator, X, y, 
                                                                     train_sizes=train_sizes, 
                                                                     cv=cv_strategy, scoring='log_loss')
    plt.figure(figsize=(10, 6), dpi=90)
    plt.plot(train_sizes, abs(train_scores.mean(axis = 1)), marker='.', label='Train data error', color='r')
    plt.plot(train_sizes, abs(test_scores.mean(axis = 1)), marker='.', label='Cross validation error', color='g')
    plt.grid(True)
    plt.xlabel('Train data sizes')
    plt.ylabel('Error')
    plt.xlim(0, 3500)
    plt.title('Train error vs Cross validation error')
    plt.legend(loc='lower right')
```

## 6) Confusion matrix and threshhold tuning

#### * Confusion matrix for (LFP prediction)

A confusion matrix is a table that is used to describe the performance of a classification model on a set of test data for which the true values are known.

Confusion matrix elements defenition:
- True Positives (TP): These are cases in which we predicted yes (participated).
- True Negatives (TN): We predicted no, and customers do not default.
- False Positives (FP): We predicted did not participated, but customers actually participated (" Type I error ").
- False Negatives (FN): We predicted participated, but customers actually did not participated (" Type II error ").

Let's plot the confusion matrix illustration the results of the Recall and Precision results

```python
def plot_confusion_matrix (cm, classes, title='Confusion matrix', cmap=plt.cm.GnBu):
    "The function prints and plots the confusion matrix"

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "red")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# Print the confusion matrix of TP, FP, FN and TN values:
cnf_matrix = metrics.confusion_matrix(y_test, gbc_y_predict)
plt.figure(figsize=(10, 6), dpi=90)
plot_confusion_matrix(cnf_matrix, classes=['Did not participated', 'Participated'],
                      title='Confusion matrix')

cnf_matrix=pd.DataFrame(cnf_matrix)

print('From the total of', y_test.shape[0],'the classifier predicted "Did not participated"', cnf_matrix.iloc[:,0].sum(),
      'times and "Participated"',cnf_matrix.iloc[:,1].sum(),'times.\n')
print ('Compare to the true results, where', cnf_matrix.iloc[1,1], 'participated and', cnf_matrix.iloc[0,0],
       'did not participated.')
```

#### * ROC curve

The ROC curve is created by plotting the true positive rate (TPR) against the false positive rate (FPR) at various threshold settings.
Each point on the ROC curve represents a sensitivity/specificity pair corresponding to a particular decision threshold. The area under the ROC curve (AUC) is a measure of how well a parameter can distinguish between two predicted classes.

## 7) Effort to improve the results-Stacking implementation

Stacking is an ensemble learning technique to combine multiple classification models via a meta-classifier. The individual classification models are trained based on the complete training set; then, the meta-classifier is fitted based on the outputs -- meta-features -- of the individual classification models in the ensemble. The meta-classifier can either be trained on the predicted class labels or probabilities from the ensemble

<center>
<img src="Picture2.PNG">

**Stacking - Step by step explanation:**
1. We train various machine learning algorithms (regressors or classifiers) in dataset A

2. We make predictions for each one of the algorithms for datasets B and C and we create new datasets B1 and C1 that contain only these predictions. So if we ran 10 models then B1 and C1 have 10 columns each.

3. We train a new machine learning algorithm (often referred to as Meta learner or Super learner) using B1

4. We make predictions using the Meta learner on C1

```python
from sklearn.base import BaseEstimator, ClassifierMixin
class stacking_model (BaseEstimator, ClassifierMixin):
    
    def __init__(self, model, meta_model):
        self.model=model
        self.meta_model=meta_model
        self.n=len(model)
        self.valid=None
        
    def fit(self, X, y, p=0.3, cv=3, random_state=30):
        train_X, valid, train_y, y_valid = model_selection.train_test_split(X, y, test_size=p, random_state=30)
        
        if (p>0):
            self.valid=np.zeros((valid.shape[0], self.n))
            for i, model in enumerate(self.model):
                model.fit(train_X, train_y)
                self.valid[:,i]=model.predict(valid)
                
            self.meta_model.fit(self.valid, y_valid)
            
        else:
            self.valid = np.zeros((X.shape[0], self.n))
            for t, clf in enumerate(self.model):
                self.valid[:, t] += model_selection.cross_val_predict(clf, X, y, cv=cv,  method='predict', n_jobs=-1)
                clf.fit(X, y)
            self.meta_model.fit(self.valid, y)  
        return self

    def predict(self, X, y=None):
        X_meta = np.zeros((X.shape[0], self.n))
        
        for t, clf in enumerate(self.model):
            X_meta[:, t] = clf.predict(X)
        
        results = self.meta_model.predict(X_meta)
        
        return (results)
```
